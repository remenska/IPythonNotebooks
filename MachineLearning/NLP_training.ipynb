```python
def review_to_words(raw_review):
    # Function to convert a raw review to a string of words
    # The input is a single string (a raw movie review), and 
    # the output is a single string (a preprocessed movie review)
    #
    # 1. Remove HTML
    review_text = BeautifulSoup(raw_review).get_text()
    # 2. Remove non-letters       
    letters_only = re.sub("[^a-zA-Z]", " ", review_text)
    # 3. convert to lower case, split into words
    words = letters_only.lower().split()
    # 4. searching a set is much faster than searching a list, so convert to set first
    stops = set(stopwords.words("english"))
    # 5. remove stopwords
    meaningful_words = [w for w in words if not w in stops]
    # 6. join the words into one string separated by space
    return(" ".join(meaningful_words))
```

```python
clean_review = review_to_words(train["review"][0])
```

```python
clean_review
```

```python
# get number of reviews based on the dataframe "review" column size
num_reviews = train["review"].size
num_reviews
```

```python
# initialize empty list to hold the clean reviews 
clean_train_reviews = []
for i in xrange(0, num_reviews):
    # Call our function for each one, and add the result to the list of clean reviews
    if( (i+1)%1000 == 0):
        print "Review %d of %d\n" % (i+1, num_reviews)
    clean_train_reviews.append(review_to_words(train["review"][i]))
```

### Creating Features from a Bag of Words (Using scikit-learn)

The Bag of Words model learns a vocabulary from all of the documents, then models each document by counting the number of times each word appears. In the IMDB data, we have a very large number of reviews, which will give us a large vocabulary. To limit the size of the feature vectors, we should choose some maximum vocabulary size. Below, we use the 5000 most frequent words (remembering that stop words have already been removed).

```python
# using the feature_extraction module from scikit-learn to create bag-of-words features
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer = "word", 
                             tokenizer = None, 
                             preprocessor = None,
                             stop_words = None, 
                             max_features = 5000
                            )
```

```python
train_data_features = vectorizer.fit_transform(clean_train_reviews) # takes some time
```

```python
# convert the result to a numpy array
train_data_features = train_data_features.toarray()
```

```python
train_data_features.shape
```

```python
# take a look at the words in the vocabulary
vocab = vectorizer.get_feature_names()
vocab
```

```python
# count words in the vocabulary
import numpy as np
dist = np.sum(train_data_features, axis = 0)
```

### Random Forest

The Random Forest algorithm is included in scikit-learn (Random Forest uses many tree-based classifiers to make predictions, hence the "forest"). Below, we set the number of trees to 100 as a reasonable default value.

```python
print "Training the random forest..."
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators = 100)
forest = forest.fit(train_data_features, train["sentiment"]) # takes some time
```

```python
# run the trained Random Forest on our test set and create a submission file
# our task is to predict the sentiment label
```

```python
test = pd.read_csv("testData.tsv", header=0, delimiter="\t", quoting=3)
test.shape
```

```python
num_reviews = len(test["review"])
clean_test_reviews = []
print "Cleaning and parsing the test set movie reviews ... \n"
for i in xrange(0, num_reviews):
    if((i+1)%1000  ==0):
        print "Review %d of %d\n" %(i+1, num_reviews)
    clean_review = review_to_words(test["review"][i])
    clean_test_reviews.append(clean_review)
```

```python
# get the bag of words for the test set, convert it to numpy array
test_data_features = vectorizer.transform(clean_test_reviews)
test_data_features = test_data_features.toarray()
```

```python
# use RandomForest to make sentiment label predictions
result = forest.predict(test_data_features)
# copy results to Pandas dataframes with an "id" column and "sentiment" one
output = pd.DataFrame(data = {"id":test["id"], "sentiment":result})
```

```python
# Pandas for CVS output
output.to_csv("Bag_of_Words_model.csv", index=False, quoting=3 )
```
